```python
# Importing libraries
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass # To get the password without showing the input
password = getpass.getpass()
```

```python
# Create a Python connection with SQL database
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
engine = create_engine(connection_string)
```

```python
# Importing rentals_may table from SQL
rentals_may = pd.read_sql_query("SELECT * FROM SAKILA.RENTALS_MAY", engine)
rentals_may
```

```python
# Importing rentals_june table from SQL
rentals_june = pd.read_sql_query("SELECT * FROM SAKILA.RENTALS_JUNE", engine)
rentals_june
```

```python
# Checking the number of rentals for each customer for May
query_may = "select customer_id, count(*) as number_of_rentals \
from sakila.rentals_may \
group by customer_id"

number_of_rentals_may = pd.read_sql_query(query_may, engine)
number_of_rentals_may
```

```python
# Checking the number of rentals for each customer for June
query_june = "select customer_id, count(*) as number_of_rentals \
from sakila.rentals_june \
group by customer_id"

number_of_rentals_june = pd.read_sql_query(query_june, engine)
number_of_rentals_june
```

```python
# Writing a function that checks if customer borrowed more or less films in the month of June as compared to May
def check_customer(customer_id):
    if number_of_rentals_june[number_of_rentals_june["customer_id"] == customer_id].iloc[0,1] > number_of_rentals_may[number_of_rentals_may["customer_id"] == customer_id].iloc[0,1]:
        print("Customer borrowed more films in June than May")
    elif number_of_rentals_june[number_of_rentals_june["customer_id"] == customer_id].iloc[0,1] < number_of_rentals_may[number_of_rentals_may["customer_id"] == customer_id].iloc[0,1]:
        print("Customer borrowed more films in May than June")
    else:
        print("Customer borrowed the same amount of films in May and June")
```

```python
# Applying the function to the customer_id == 100
check_customer(100)
```

```python
# Applying the function to the customer_id == 150
check_customer(150)
```

```python
# Applying the function to the customer_id == 70
check_customer(70)
```
