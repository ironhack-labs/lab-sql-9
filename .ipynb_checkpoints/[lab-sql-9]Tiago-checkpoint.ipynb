```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
```

```python
import getpass 
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
engine = create_engine(connection_string)
may = pd.read_sql_query("select concat(c.first_name, ' ', c.last_name) as name, count(r.rental_id) as qty \
from sakila.rentals_may as r \
left join sakila.customer as c \
on  c.customer_id = r.customer_id \
group by r.customer_id \
order by 2 desc;", engine)
june = pd.read_sql_query("select concat(c.first_name, ' ', c.last_name) as name, count(r.rental_id) as qty \
from sakila.rentals_june as r \
left join sakila.customer as c \
on  c.customer_id = r.customer_id \
group by r.customer_id \
order by 2 desc;", engine)
```

```python
df_all= pd.merge(may, june, how='inner', on='name')
```

```python
df_all=df_all.rename(columns={"qty_x": "may", "qty_y": "june"})
```

```python
df_all
```

```python
may= df_all["may"].values.tolist()
june= df_all["june"].values.tolist()
```

```python


may_count = 0
june_count = 0
    
for i in range(0, len(may)):
    if (may[i] > june[i]):
        may_count += 1
    elif (may[i] < june[i]):
        june_count += 1
    else: pass

print("May Count:", may_count)
print("June Count:", june_count)
```
