```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass

password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
query = """
        SELECT
              customer_id,
              COUNT(*) as num_rentals
        FROM sakila.rentals_may
        GROUP BY customer_id;
        """
pd.read_sql_query(query,engine)
```

```python
query = """
        SELECT
              customer_id,
              COUNT(*) as num_rentals
        FROM sakila.rentals_june
        GROUP BY customer_id;
        """
pd.read_sql_query(query,engine)
```

```python
def compare_monthly_borrowing(customer_id, may_borrowed_films, june_borrowed_films):
    if june_borrowed_films > may_borrowed_films:
        return f"Customer {customer_id} borrowed more films in June ({june_borrowed_films}) compared to May ({may_borrowed_films})."
    elif june_borrowed_films < may_borrowed_films:
        return f"Customer {customer_id} borrowed fewer films in June ({june_borrowed_films}) compared to May ({may_borrowed_films})."
    else:
        return f"Customer {customer_id} borrowed the same number of films in May ({may_borrowed_films}) and June ({june_borrowed_films})."
```
