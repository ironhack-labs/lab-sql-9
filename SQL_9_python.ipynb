-- 6.Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

```python
def compare_rentals(df, column1, column2):
    result = []

    for index, row in df.iterrows():
        if row[column1] > row[column2]:
            result.append(f"{column1} is higher")
        elif row[column1] < row[column2]:
            result.append(f"{column2} is higher")
        else:
            result.append("Rentals in May and June are the same")

    df['rentals_dif'] = result
    return df
```

```python
compare_rentals (rentals_may_june, 'May_rentals', 'June_rentals')
```
