```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
engine = create_engine(connection_string)
rentals_june = pd.read_sql_query('select * from sakila.rentals_june', engine)
```

```python
# Check the number of rentals for each customer for May

num_rentals_may = rentals_may[['customer_id', 'first_name', 'last_name']].value_counts().reset_index()
num_rentals_may.columns = ['customer_id', 'first_name', 'last_name', 'num_rentals_may']
num_rentals_may
```

```python
# Check the number of rentals for each customer for June

num_rentals_june = rentals_june[['customer_id', 'first_name', 'last_name']].value_counts().reset_index()
num_rentals_june.columns = ['customer_id', 'first_name', 'last_name', 'num_rentals_june']
num_rentals_june
```

```python
# Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

# Hint: For this part, you can create a join between the two dataframes created before, using the merge function 
# available for pandas dataframes. Here is a link to the documentation for the merge function.
```

```python
rentals_merged = pd.merge(num_rentals_may, num_rentals_june)
rentals_merged
```

```python
def more_or_less(df):
    df['rental_diff'] = df['num_rentals_june'] - df['num_rentals_may']
    return df
```
