```python
#  7.1 Check the number of rentals for each customer for May

data_cus_may = pd.read_sql_query('select rent.customer_id, customer.first_name, customer.last_name,count(rent.rental_id) as rentals_may \
                                  from sakila.rentals_may as rent \
                                  join sakila.customer as customer \
                                  on rent.customer_id = customer.customer_id \
                                  group by customer_id \
                                  order by customer_id', engine)
data_cus_may
```

```python
# 7.2 Check the number of rentals for each customer for June

data_cus_june = pd.read_sql_query('select rent.customer_id, customer.first_name, customer.last_name,count(rent.rental_id) as rentals_june \
                                  from sakila.rentals_june as rent \
                                  join sakila.customer as customer \
                                  on rent.customer_id = customer.customer_id \
                                  group by customer_id \
                                  order by customer_id', engine)
data_cus_june
```

```python
# 8. Write a function that checks if customer borrowed more or less films in the month of June as compared to May.
```

```python
# Creating the data frame
data_may_june = pd.merge(data_cus_may, data_cus_june, on = ['customer_id', 'first_name', 'last_name'], how = 'outer')

data_may_june= data_may_june.fillna(0)
data_may_june
```

```python
# Function for comparing each customer rentals between May and June

def compare_rentals(row):
    may_rentals = row['rentals_may']
    june_rentals = row['rentals_june']

    if may_rentals > june_rentals:
        return 'Less rentals in June'
    elif may_rentals < june_rentals:
        return 'More rentals in June'
    else:
        return 'Equal rentals in May and June'


data_may_june['rental_comparison'] = data_may_june.apply(compare_rentals, axis=1)
data_may_june['total_rentals'] = (data_may_june['rentals_may'] + data_may_june['rentals_june'])
data_may_june
```

```python
# Function for comparing total rentals between May and June

total_may = data_may_june['rentals_may'].sum()
total_june = data_may_june['rentals_june'].sum()

if total_may > total_june:
    print('There where more rentals in May - in total: ', total_may, 'compared to ', total_june, 'rentals')
elif total_may < total_june:
    print('There where more rentals in June - in total: ', total_june, 'compared to ', total_may, 'rentals')
else:
    print('There where the same aount of rentals in May and June - in total: ', total_june)
```
