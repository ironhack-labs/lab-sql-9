```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
engine = create_engine(connection_string)
data = pd.read_sql_query('select * from sakila.rentals_may_june', engine)
data = data.set_index('customer_id')
data
```

```python
def compare_rentals_may_june(df):
    df['most_rentals'] = np.where(df['nr_rentals_may'] > df['nr_rentals_june'], 'may',
                              np.where(df['nr_rentals_may'] < df['nr_rentals_june'], 'june', 'equal'))
    return df
```

```python
compare_rentals_may_june(data)
```
