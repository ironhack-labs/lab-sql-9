```python
# Check the number of rentals for each customer for May using Python


df_rental_may.groupby('customer_id')['rental_id'].count()

# Hint: You can store the results from the two queries in two separate dataframes.

df_may_count = df_rental_may.groupby('customer_id')['rental_id'].count()

df_may_count
```

```python
# Check the number of rentals for each customer for Juen
# same techniq

df_rental_june.groupby('customer_id')['rental_id'].count()

# Hint: You can store the results from the two queries in two separate dataframes.

df_june_count = df_rental_june.groupby('customer_id')['rental_id'].count()

df_june_count
```

```python
# Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

# Hint: For this part, you can create a join between the two dataframes created before, using the merge function available for pandas dataframes. Here is a link to the documentation for the merge function.
df_may_count = pd.DataFrame(df_may_count)


merged_df = df_may_count.merge(df_june_count, on='customer_id', suffixes=('_may', '_june'), how='left')

merged_df
```

```python

merged_df['rental_total'] = merged_df['rental_id_may'] - merged_df['rental_id_june']

merged_df
```

```python
# creating New column to evaluate if month of may got more rental than June month


merged_df['rental_total'] = merged_df['rental_id_may'] - merged_df['rental_id_june']

Borrow_month = []


count_june = 0
count_may = 0

# function to define wich month customer did the most rental
def compare_rental_counts(row):

    global count_june, count_may

    if row ['rental_total'] < 0:
        Borrow_month.append( 'June')
        # count  location total for June month
        count_june += 1          
    else :
        Borrow_month.append('May')
         # count  location total for May mont
        count_may += 1
            


merged_df.apply(compare_rental_counts, axis=1)

# choose whcih month had the more rental
if count_june > count_may :
    print(f"Count of June is the bigger: {count_june} rental")

if count_june < count_may :
    print(f"Count of May is the bigger with: {count_may} rental")

merged_df['comparison_result'] = Borrow_month
```
